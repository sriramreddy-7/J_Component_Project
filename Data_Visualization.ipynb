```python
import matplotlib.pyplot as plt
import os
import cv2
```

```python
data_dir = '/content/drive/MyDrive/J_CompAIML/DATASET/BTDS'
classes = ['BrainTumor', 'Healthy']
```

```python
rows, cols = 4, 4

fig, ax = plt.subplots(rows, cols, figsize=(10, 10))
ax = ax.ravel()

for i, label in enumerate(classes):
    path = os.path.join(data_dir, label)
    files = os.listdir(path)

    for j in range(cols):
        img_path = os.path.join(path, files[j])
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        ax[i * cols + j].imshow(img)
        ax[i * cols + j].set_title(label)
        ax[i * cols + j].axis('off')

plt.tight_layout()
plt.show()
```

```python
num_images = []
for label in classes:
    path = os.path.join(data_dir, label)
    num_images.append(len(os.listdir(path)))

plt.bar(classes, num_images)
plt.title('Number of Images in Each Class')
plt.xlabel('Class')
plt.ylabel('Number of Images')
plt.show()
```

```python
import os
import cv2

# Define the dataset directory and classes
data_dir = '/content/drive/MyDrive/J_CompAIML/DATASET/BTDS'
classes = ['BrainTumor', 'Healthy']
flag=0
# Loop over the classes
for cls in classes:
    # Get the file names for the class
    files = os.listdir(os.path.join(data_dir, cls))
    # Loop over the files
    for file in files:
        # Load the image
        img_path = os.path.join(data_dir, cls, file)
        img = cv2.imread(img_path)
        # Check if the image is None (i.e., could not be read)
        if img is None:
            print(f"Removing {img_path}")
            # Remove the file
            os.remove(img_path)
            flag+=1
```
